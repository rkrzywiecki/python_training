```r
data = runif(20, 1, 10)
data
```

```r
mean = mean(data)
```

```r
print(mean)
```

```r
median = median(data)
median
```

```r
mode <- function(x) {
    ux <- unique(x)
    ux[which.max(tabulate(match(x, ux)))]
}
result <- mode(data)
result
```

```r
variance <- var(data)
variance
```

```r
sd <- sqrt(var(data))
sd
```
